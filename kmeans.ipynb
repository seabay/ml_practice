```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
class Kmeans:

  def __init__(self, k):
    self.k=k


  def fit(self, data):

    row=data.shape[0]
    col=data.shape[1]

    centre=np.random.randn(self.k, col)
    cluster=np.zeros(data.shape[0])
    old_centre=np.random.randn(self.k, col)
    for loop in range(10):
      ### update cluster
      for i in range(row):
        dist=float('+inf')
        f_j=None
        for j in range(self.k):
          d_i_j = np.linalg.norm(data[i] - centre[j])
          if dist>d_i_j:
            dist=d_i_j
            f_j=j
        
        cluster[i]=f_j
      
      plot.scatter(data[:, 0], data[:, 1], c=cluster)
      plot.scatter(centre[:, 0], centre[:, 1], marker='*')
      plot.show()

      ### update centres

      for j in range(self.k):
        mask=(cluster==j)
        centre[j]=np.mean(data[mask], axis=0)

    
      ### stop

      #print(old_centre-centre)
      diff=np.mean(np.linalg.norm(old_centre-centre, axis=1))
      print(diff)
      if diff<1.0e-4:
        break

      
      old_centre=centre.copy()

```

```python
x1=np.random.randn(30, 2)+np.array([2, 3])
x2=np.random.randn(30, 2)+np.array([10, -5])
x3=np.random.randn(30, 2)+np.array([-5, -10])
```

```python
data=np.concatenate([x1, x2, x3], axis=0)
```

```python
plot.scatter(data[:, 0], data[:, 1])
plot.show()
```

```python
k=Kmeans(3)
```

```python
k.fit(data)
```
