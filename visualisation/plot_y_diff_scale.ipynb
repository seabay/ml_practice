```
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('Sheet2.csv', names=['date', 'transaction_size', 'customer_size'])
```

```
df['transaction_size']=df['transaction_size'].astype('int32')
```

```
df.head(2)
```

```

ax=df[['date', 'transaction_size']].plot(x='date', kind='bar', width = 0.5, color='b', figsize=(15,10))
df[['date', 'customer_size']].plot(x='date', linestyle='-', marker='o', c='r', ax=ax)

axes = plt.gca()
axes.set_ylim([10000, 2.5e7])
ax.set_xticklabels(df['date'], rotation=90, horizontalalignment='right')

for i,j in zip([x for x in range(0, 20, 1)],df['transaction_size']):
    ax.annotate(str(j),xy=(i+0.5,j+500000), rotation=50, horizontalalignment='right')

for i,j in zip([x for x in range(0, 20, 1)],df['customer_size']):
    ax.annotate(str(j),xy=(i+0.8,j), rotation=50, horizontalalignment='right')

plt.show()
```

```
plt.figure(figsize=(20,10))

fig,ax = plt.subplots()
# make a plot
ax.plot(df.date, df.customer_size, color="red", marker="o")
# set x-axis label
ax.set_xlabel("date",fontsize=14)
# set y-axis label
ax.set_ylabel("customer_size",color="red",fontsize=14)
ax.set_xticklabels(df['date'], rotation=90, horizontalalignment='right')
#ax.grid(axis='y')

ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.bar(df.date, df.transaction_size,color="blue", width=0.1)
ax2.set_ylabel("transaction_size",color="blue",fontsize=14)
#ax2.grid(axis='y')

for i,j in zip([x for x in range(0, 20, 1)],df['transaction_size']):
    ax2.annotate(str(j),xy=(i+0.6,j), rotation=50, horizontalalignment='right')

for i,j in zip([x for x in range(0, 20, 1)],df['customer_size']):
    ax.annotate(str(j),xy=(i+0.3,j), rotation=50, horizontalalignment='right')

fig.set_size_inches(15, 10)
plt.show()
```
